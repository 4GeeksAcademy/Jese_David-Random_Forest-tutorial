# **RANDOM FOREST**

```python
#LIBRERIAS
import pandas as pd
import matplotlib.pyplot as plt
from pickle import dump

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

from sklearn.model_selection import GridSearchCV
```

```python
#DATA
sel_train= pd.read_csv ("/workspaces/Jese_David-Random_Forest-tutorial/Jese_David-Decision-Tree-project/data/Datasets/sel_dataset_train.csv")
sel_test = pd.read_csv ("/workspaces/Jese_David-Random_Forest-tutorial/Jese_David-Decision-Tree-project/data/Datasets/sel_dataset_test.csv")
target = "Outcome"

#Split
X_sel_train = sel_train.drop(columns=[target])
y_sel_train = sel_train[target]

X_sel_test = sel_test.drop(columns=[target])
y_sel_test = sel_test[target]
```

### **RANDOM FOREST BASE**

```python
rf_sel = RandomForestClassifier(
    n_estimators=200,        # nº de árboles
    max_depth=None,          # sin limitar de momento
    random_state=42,
    class_weight="balanced", # por el desbalance de Outcome
    n_jobs=-1                # usa todos los cores
)

rf_sel.fit(X_sel_train, y_sel_train)

y_sel_pred_rf = rf_sel.predict(X_sel_test)

print("\n=== RANDOM FOREST (SEL) ===")
print("Accuracy test:", accuracy_score(y_sel_test, y_sel_pred_rf))
print("\nMatriz de confusión:\n", confusion_matrix(y_sel_test, y_sel_pred_rf))
print("\nReporte de clasificación:\n", classification_report(y_sel_test, y_sel_pred_rf))
```

### **Guardo Modelo**

```python
dump(
    rf_sel,
    open("../models/RandomForest_sel.sav", "wb")
)
```
